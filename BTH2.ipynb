```python
import pandas as pd
dt = pd.read_xml('D:\python\PythonNangCao\AnyConv.com__netflix_data (1).xml')

# Hiển thị ra màn hình
print(dt.head())
```

```python
from lxml import etree

tree = etree.parse(r"D:\python\PythonNangCao\AnyConv.com__netflix_data (1).xml")

# 1. Các bộ phim (Movie) phát hành năm 2020
# Giả sử cấu trúc là <record><type>Movie</type><release_year>2020</release_year></record>
movies_2020 = tree.xpath("//record[type='Movie' and release_year='2020']")

# 2. TV Show sản xuất tại United States phát hành năm 2021
tv_shows_us_2021 = tree.xpath("//record[type='TV Show' and country='United States' and release_year='2021']")
```

```python
# 1. Số bộ phim sản xuất tại United States vào năm 2009
count_us_2009 = len(dt[(dt['type'] == 'Movie') & (dt['country'] == 'United States') & (dt['release_year'] == 2009)])
print(f"Số phim US năm 2009: {count_us_2009}")

# 2. Số video clips được sản xuất theo các nước
stats_country = dt['country'].value_counts()
print(stats_country)

# 3. Các chương trình "TV Show" có từ 2 mùa trở lên
# Lưu ý: Cột duration của TV Show thường ghi là "2 Seasons", ta cần tách lấy số
tv_shows = dt[dt['type'] == 'TV Show'].copy()
# Tách lấy phần ố và chuyển sang kiểu int
tv_shows['seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(int)
tv_2_seasons = tv_shows[tv_shows['seasons'] >= 2]
print(tv_2_seasons[['title', 'duration']])
```

```python
movies = dt[dt['type'] == 'Movie'].copy()
movies['duration_num'] = movies['duration'].str.extract('(\d+)').astype(float)

short_movies = movies[movies['duration_num'] <= 120]

short_movies.to_xml('ShortVideo.xml', index=False, encoding='utf-8')
print("Đã xuất file ShortVideo.xml thành công!")
```
